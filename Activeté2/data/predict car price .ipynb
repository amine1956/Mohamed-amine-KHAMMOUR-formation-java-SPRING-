## Numerical variables 

1. So the first probleme in Fiscal power is the string 'CV ' that we find in some rows 

```python
df['Fiscal power']=df['Fiscal power'].astype(str)
```

```python
#remove CV dans le variable Fiscal power
def removesuffix(input_string, suffix):
    if suffix and input_string.endswith(suffix):
        return input_string[:-len(suffix)]
    return input_string
```

```python
#remove plus de dans Fiscal power
def removepreffix(input_string, preffix):
    if preffix and input_string.startswith(preffix):
        return input_string[len(preffix):]
    return input_string
```

```python
df['Fiscal power']=df['Fiscal power'].apply(lambda text: removesuffix(text,'CV'))

df['Fiscal power']=df['Fiscal power'].apply(lambda text: removepreffix(text,'Plus de '))
```

```python
df['Fiscal power']=df['Fiscal power'].astype(float)
```

```python
df.dtypes
```

```python
#elver le nan
avg_Fiscal_power = int(df["Fiscal power"].mean(axis=0))
print("Average of Fiscal power:", avg_Fiscal_power)
df["Fiscal power"].replace(np.nan, avg_Fiscal_power, inplace=True)
```

2. Now we need to find a solution to mileage and transform it to one value insted of interval 

```python

df['Mileage']=df['Mileage'].astype(str)
df['Mileage']
```

```python
#les valeurs non null
var_mean=df['Mileage'].dropna(axis=0)
var_mean
```

```python
df['Mileage'].isna().sum()
```

```python
def trans_to_mean(valeur):
    if "Plusde" in valeur :
        return (valeur.replace('Plusde',' '))
    if 'na' in valeur :
        return 0
    return (float(valeur[:valeur.find("-")])+float(valeur[valeur.find("-")+1:]))/2
```

```python
#apply prcouritre ling par line 
var_mean=var_mean.apply(lambda val: trans_to_mean(val))
var_mean
```

```python
var_mean=var_mean.astype(float)
```

```python
var_mean=var_mean.mean()
```

```python
df['Mileage'].replace(np.nan,var_mean, inplace=True)
```

```python
df['Mileage']=df['Mileage'].apply(lambda val: trans_to_mean(val))
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
impute_data=df[["Fuel type","Brand","Model","Boite à vitesse","Origine","Première main","Sector","Nombre de portes"]]
```

```python
impute_data
impute_data['Model']=impute_data['Model'].astype(str)
impute_data['Sector']=impute_data['Sector'].astype(str)
impute_data['Nombre de portes']=impute_data['Nombre de portes'].astype(str)
impute_data
```

```python
encoder = OrdinalEncoder()
imputer = KNNImputer()
# create a list of categorical columns to iterate over
cat_cols = ["Fuel type","Brand","Model","Boite à vitesse","Origine","Première main","Sector","Nombre de portes"]
def encode(data):
    '''function to encode non-null data and replace it in the original data'''
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the dat a for encoding
    impute_reshape = nonulls.reshape(-1,1)
    #encode date
    impute_ordinal = encoder.fit_transform(impute_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data
#create a for loop to iterate through each column in the data
for columns in cat_cols:
    encode(impute_data[columns])
impute_data
```

```python
col=["Fuel type","Brand","Model","Boite à vitesse","Origine","Première main","Sector","Nombre de portes"]
```

```python
encode_data = pd.DataFrame(np.round(imputer.fit_transform(impute_data)),columns = col)
```
